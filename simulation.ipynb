#### Probability function

```python
def prob(mu, *args):
    total = 0.0
    for arg in args:
        if arg[0] == 1:
            total += np.exp(arg[1] * mu)
    prob = []
    for arg in args:
        if arg[0] == 1:
            prob.append(np.exp(mu*arg[1]) / total)
        else:
            prob.append(0)
    return prob
```

#### Simulation

```python
def sim(probabilityVector):
    r = np.random.random()
    n = 0
    for i,value in enumerate(probabilityVector):
        n += value
        if r <= n:
            break
    return i
```
